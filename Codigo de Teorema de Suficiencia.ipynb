```python
#Definimos nuestra variable a usar
import sympy as sp

x = sp.symbols('x')

# Escribe aqui la funcion como texto
entrada = x**4 / ((x - 1)*(x - 3)**3)

# Convertimos el texto a funcion simbolica
f = sp.sympify(entrada)

print("Funcion que se ingreso:", f)
```

```python
# Primera derivada
f_prime = sp.diff(f, x)
f_prime = sp.simplify(f_prime)

# Se halla el dominio de la funcion
den = sp.denom(f)
puntos_no_dominio = sp.solve(den, x)

print("Funcion f(x) =", f)
print("Primera derivada f'(x) =", f_prime)

if puntos_no_dominio:
    print("La funcion NO existe en:", puntos_no_dominio)
else:
    print("La funcion existe en los reales.")

# Dominio de la derivada
den_der = sp.denom(f_prime)
puntos_no_derivada = sp.solve(den_der, x)

if puntos_no_derivada:
    print("La derivada NO existe en:", puntos_no_derivada)
else:
    print("La derivada existe en los reales.")
```

```python
# Hallamos los puntos criticos
puntos_criticos = sp.solve(f_prime, x)

print("Los puntos criticos son:", puntos_criticos)
print("\nAplicamos el teorema de suficiencia:\n")

for pc in puntos_criticos:
    print("======================================")
    print(f"Analizando el punto critico x* = {pc}\n")
    print("Derivamos y evaluamos")
    print(" ")

    #  Verificamos que la funcion exista en el punto
    if f.subs(x, pc) in [sp.oo, -sp.oo, sp.nan]:
        print("La funcion no existe en este punto por lo que NO es punto critico valido.\n")
        continue

    orden = 2
    derivada = sp.diff(f_prime, x)

    while True:

        derivada = sp.simplify(derivada)
        valor = sp.simplify(derivada.subs(x, pc))

        print(f"Derivada de orden {orden}:")
        print(f"f^({orden})(x) = {derivada}")
        print(f"f^({orden})({pc}) = {valor}")

        if not valor.equals(0):
            print("\nLa derivada es diferente de cero.")

            if orden % 2 == 0:
                if valor > 0:
                    m = "n es par y positivo por lo tanto es un MINIMO local"
                else:
                    m = "n es par y negativo por lo tanto es un MAXIMO local"
            else:
                m = "n es impar por lo tanto es un Punto de inflexion"

            break

        else:
            print("Aun no nos dice nada, seguimos derivando.\n")
            derivada = sp.diff(derivada, x)
            orden += 1

    # Evaluamos en la funcion
    pc_pi = sp.nsimplify(pc, [sp.pi])
    valor_funcion = sp.nsimplify(f.subs(x, pc), [sp.pi])

    print(f"Note que {m}")
    print(f"y lo tiene en f({pc}) = {valor_funcion}")
    print(f"en el punto ({pc}, {valor_funcion})")
    print("======================================\n")
```
